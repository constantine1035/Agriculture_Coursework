```python
import os

def list_files(directory, output_file):
    try:
        with open(output_file, 'w', encoding='utf-8') as f:
            for file_name in os.listdir(directory):
                file_path = os.path.join(directory, file_name)
                if os.path.isfile(file_path):
                    f.write(file_name + '\n')
        print(f'Ð¤Ð°Ð¹Ð»Ñ‹ Ð·Ð°Ð¿Ð¸ÑÐ°Ð½Ñ‹ Ð² {output_file}')
    except Exception as e:
        print(f'ÐžÑˆÐ¸Ð±ÐºÐ°: {e}')

list_files("./AgricultureVisionYOLOrgb/val/images", "./datasetinfo/val_images.txt")
list_files("./AgricultureVisionYOLOrgb/val/labels", "./datasetinfo/val_labels.txt")
```

```python
import pandas as pd
import numpy as np
from ultralytics import YOLO
import torch

import shutil
import os

import random

from PIL import Image
import cv2
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
device = torch.device("cuda")
```

```python
with open("dataset.yaml", "w") as f:
    f.write("""path: ./AgricultureVisionYOLOrgb
names:
- double_plant
- drydown
- endrow
- nutrient_deficiency
- planter_skip
- storm_damage
- water
- waterway
- weed_cluster
nc: 9
train: train/images
val: val/images""")
```

```python
model_n = YOLO("yolo11n-seg.pt")
model_n.train(data='dataset.yaml',
              epochs=25,
              imgsz=512,
              optimizer = 'AdamW',
              lr0 = 1e-2, 
              project = 'YOLOSEG',
              name='YOLOn',
              batch=64,
              device=device,
              seed=42,
              patience=5)
```
